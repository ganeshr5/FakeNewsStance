```python
# calculate polarity and subjecivity using textblob
from textblob import TextBlob

#TextBlob("not a very great calculation").sentiment
```

```python
sent_train_title1 = [] 
sent_train_title1_len = []
sent_train_title1_polarity = []
sent_train_title1_subjectivity = []
sent_train_title2 = [] 
sent_train_title2_len = []
sent_train_title2_polarity = []
sent_train_title2_subjectivity = []
```

```python
for id in range(0, len(df_train)):
  sent_train_title1.append(df_train.title1_en[id])
  sent_train_title1_len.append(len(df_train.title1_en[id]))
  sent_train_title1_polarity.append(TextBlob(df_train.title1_en[id]).sentiment.polarity)
  sent_train_title1_subjectivity.append(TextBlob(df_train.title1_en[id]).sentiment.subjectivity)
  sent_train_title2.append(df_train.title2_en[id])
  sent_train_title2_len.append(len(df_train.title2_en[id]))
  sent_train_title2_polarity.append(TextBlob(df_train.title2_en[id]).sentiment.polarity)
  sent_train_title2_subjectivity.append(TextBlob(df_train.title2_en[id]).sentiment.subjectivity)
```

```python
senti_train = pd.DataFrame(columns = ['title1', 'title1_len', 'title1_polarity', 'title1_subjectivity', 'title2', 'title2_len', 'title2_polarity', 'title2_subjectivity'])
```

```python
senti_train['title1'] = pd.Series(sent_train_title1).values
senti_train['title1_len'] = pd.Series(sent_train_title1_len).values
senti_train['title1_polarity'] = pd.Series(sent_train_title1_polarity).values
senti_train['title1_subjectivity'] = pd.Series(sent_train_title1_subjectivity).values
senti_train['title2'] = pd.Series(sent_train_title2).values
senti_train['title2_len'] = pd.Series(sent_train_title2_len).values
senti_train['title2_polarity'] = pd.Series(sent_train_title2_polarity).values
senti_train['title2_subjectivity'] = pd.Series(sent_train_title2_subjectivity).values
```
